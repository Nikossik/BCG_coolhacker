```python
import pandas as pd
import datetime as dt
import numpy as np
import ml_metrics as metrics
from tqdm import tqdm_notebook
from catboost import CatBoostRegressor
from sklearn.preprocessing import LabelEncoder
import requests
import warnings
warnings.filterwarnings('ignore')
```

#### Preprocessing

```python
def activity(x, activity=[5,6,8,9]):
    month = int(x.split('-')[1])
    if month in activity:
        return 1
    else:
        return 0

def season(x):
    month = int(x.split('-')[1])
    if month in [12,1,2]:
        return 0
    elif month in [3,4,5]:
        return 1
    elif month in [6,7,8]:
        return 2
    else:
        return 3
```

```python
df['activity'] = df['Date'].apply(activity)
df['season'] = df['Date'].apply(season)

df['Date'] = pd.to_datetime(df['Date'])

df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['week'] = df['Date'].dt.week
df['day'] = df['Date'].dt.day
df['day_of_week'] = df['Date'].dt.dayofweek

df.index = df['Date']
df.drop('Date', axis=1, inplace=True)

df_with_regs = df.copy()

uniques = list(df['region'].unique())
def regs(x):
    return uniques.index(x)

df['region'] = df['region'].apply(regs)
df4 = df[['region', 'invitro_count', 'population']]

df.head()
```

```python
mae = []
cats = []

features = ['Вирус клещевого энцефалита(кач) РНК', 'Вирус клещевого энцефалита(кач) РНКprob', 'Borrelia miyamotoi (кач) ДНК', 'Borrelia miyamotoi (кач) ДНКprob', 'Ehrlichia muris/chaffeеnsis(кач) ДНК', 'Ehrlichia muris/chaffeеnsis(кач) ДНКprob', 'Anaplasma Phagocytophillum(кач) ДНК', 'Anaplasma Phagocytophillum(кач) ДНКprob', 'Borrelia burgdorferi s.l (кач) ДНК', 'Borrelia burgdorferi s.l (кач) ДНКprob']

for i in tqdm_notebook(range(67)):
    for j in range(len(features)):
        if np.array(datasets[i][features[j]][:'2021-05-27'].value_counts())[0] == 835:
            mae.append(metrics.mae(datasets[i][features[j]]['2021-05-27':], np.zeros(14)))
            cats.append(0)
            continue
        cats.append(CatBoostRegressor(silent=True))
        cats[len(cats)-1].fit(datasets[i][['activity', 'season', 'year', 'month', 'week', 'day', 'day_of_week', 'population', 'invitro_count', 'alltest']], datasets[i][features[j]])
        preds = cats[len(cats)-1].predict(datasets[i][['activity', 'season', 'year', 'month', 'week', 'day', 'day_of_week', 'population', 'invitro_count', 'alltest']]['2021-05-27':])

        mae.append(metrics.mae(datasets[i][features[j]]['2020-05-27':'2020-06-09'], preds))

```
